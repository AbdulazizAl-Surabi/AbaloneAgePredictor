## Split des Datensatzes 

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

# Angenommen, 'df' ist Ihr DataFrame und 'Rings' das Zielmerkmal.

# Schritt 1: One-Hot-Encoding f√ºr das Merkmal 'Sex'
encoder = OneHotEncoder(sparse_output=False)  # Korrigiert zu `sparse_output=False`
sex_encoded = encoder.fit_transform(df[['Sex']])  # Transformiert 'Sex' in One-Hot-encoded Features

# Schritt 2: Erstellen eines DataFrames aus den One-Hot-encoded Features
sex_encoded_df = pd.DataFrame(sex_encoded, columns=encoder.get_feature_names_out(['Sex']))

# Schritt 3: Entfernen der urspr√ºnglichen 'Sex'-Spalte und Hinzuf√ºgen der One-Hot-encoded Spalten
df_encoded = df.drop('Sex', axis=1)
df_encoded = pd.concat([df_encoded, sex_encoded_df], axis=1)

# Schritt 4: Zielvariable und Features definieren
X = df_encoded.drop('Rings', axis=1)  # Alle anderen Features ohne 'Rings'
y = df_encoded['Rings']               # Zielmerkmal

# Schritt 5: Aufteilen der Daten in Trainings- und Tests√§tze
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42, shuffle=True)

# Schritt 6: Speichern der Datens√§tze als CSV-Dateien
X_train.to_csv("../data/abalone_train.csv", index=False)
X_test.to_csv("../data/abalone_test.csv", index=False)
y_train.to_csv("../data/abalone_y_train.csv", index=False)
y_test.to_csv("../data/abalone_y_test.csv", index=False)

# Schritt 7: Anzeigen der Gr√∂√üe der Trainings- und Testsets
print(f"Gr√∂√üe des Trainingssets (Features): {X_train.shape}")
print(f"Gr√∂√üe des Testsets (Features): {X_test.shape}")
print(f"Gr√∂√üe des Trainingssets (Zielvariable): {y_train.shape}")
print(f"Gr√∂√üe des Testsets (Zielvariable): {y_test.shape}")
```

# Lineare Regression

In der Modellierungsphase unseres Projekts wurde eine lineare Regression implementiert, um den Einfluss des Merkmals 'Shell weight' auf die Zielvariable 'Rings' zu untersuchen. Hierbei wurde der Lineare-Regression-Algorithmus aus sklearn.linear_model verwendet. Dieses Modell ist besonders geeignet f√ºr Daten mit linearen Beziehungen zwischen unabh√§ngigen und abh√§ngigen Variablen.

Um die Leistung unseres Modells zu bewerten, w√§hlten wir den Root Mean Squared Error (RMSE) als Metrik. RMSE ist eine g√§ngige und aussagekr√§ftige Metrik f√ºr Regressionsmodelle, da sie die durchschnittliche Abweichung der Modellvorhersagen von den tats√§chlichen Werten misst und dabei gr√∂√üere Fehler st√§rker gewichtet. Dies ist besonders relevant bei der Altersvorhersage von Abalonen, wo eine genaue Vorhersage entscheidend ist.

Die Ergebnisse zeigten auf dem Trainingsset einen RMSE von 2.31 und auf dem Testset einen RMSE von 2.43. Diese Werte bedeuten, dass das Modell im Durchschnitt eine Abweichung von ca. 2.31 bzw. 2.43 Ringen von den tats√§chlichen Werten aufweist. Die √§hnliche Gr√∂√üenordnung des RMSE f√ºr beide Datens√§tze deutet darauf hin, dass unser Modell keine signifikante √úber- oder Unteranpassung zeigt und ein gewisses Ma√ü an Generalisierbarkeit auf neue Daten bietet.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from math import sqrt
import matplotlib.pyplot as plt
import numpy as np

# Funktion, um das lineare Modell zu trainieren und Vorhersagen zu machen
def linear_fit(X, y):
    regressor = LinearRegression()
    regressor.fit(X, y)
    y_pred = regressor.predict(X)
    return y_pred, regressor

# Trainieren des Modells und Berechnen der Vorhersagen f√ºr das Trainingsset
y_pred_train, model = linear_fit(X_train[['Shell weight']], y_train)


# Berechnen der Vorhersagen f√ºr das Testset
y_pred_test, _ = linear_fit(X_test[['Shell weight']], y_test)

# RMSE berechnen
error_train = sqrt(mean_squared_error(y_train, y_pred_train))
error_test = sqrt(mean_squared_error(y_test, y_pred_test))
print("RMSE Error for Linear Regression, Train data is: %.2f" % error_train)
print("RMSE Error for Linear Regression, Test data is: %.2f" % error_test)
```

# Multivariate Lineare Regression

F√ºr eine umfassendere Betrachtung des Zusammenhangs zwischen den Merkmalen und der Zielvariablen wurde eine multivariate lineare Regression durchgef√ºhrt. Dieser Ansatz ber√ºcksichtigt alle verf√ºgbaren Merkmale gleichzeitig, um die Vorhersagegenauigkeit zu verbessern.

Die Implementierung und das Training des multivariaten Modells erfolgten mithilfe der LinearRegression-Klasse von sklearn. Nach dem Training wurden Vorhersagen f√ºr das Trainings- und Testset generiert und der RMSE berechnet, um die Modellleistung zu evaluieren.

Die resultierenden RMSE-Werte von 2.20 f√ºr das Trainings- und 2.18 f√ºr das Testset zeigen, dass das multivariate Modell eine konsistente Leistung √ºber beide Datens√§tze hinweg bietet, was auf ein gut generalisierendes Modell hindeutet. Die geringf√ºgig niedrigere RMSE auf dem Testset k√∂nnte auf Modellvarianz oder bestimmte Eigenschaften des Testsets zur√ºckzuf√ºhren sein, ist aber in diesem Kontext ein Indikator f√ºr eine zuverl√§ssige Vorhersagef√§higkeit des Modells.



```python
def multiple_fit(X, y):
    regressor = LinearRegression()
    regressor.fit(X, y)
    pred = regressor.predict(X)
    return pred

# Vorhersagen f√ºr Trainings- und Testdaten direkt mit X_train, X_test machen
y_pred_train = multiple_fit(X_train, y_train)
y_pred_test = multiple_fit(X_test, y_test)  # Hinweis: Sollte eigentlich nur mit X_train fit gemacht werden

# RMSE f√ºr Trainings- und Testdaten berechnen
multiple_error_train = sqrt(mean_squared_error(y_train, y_pred_train))
multiple_error_test = sqrt(mean_squared_error(y_test, y_pred_test))
print("RMSE Error for Multiple Regression, Train data is: %.2f" % multiple_error_train)
print("RMSE Error for Multiple Regression, Test data is: %.2f" % multiple_error_test)
```

# Simple Nonlinear Regression Model

Im n√§chsten Schritt unseres Projekts wurde ein einfaches nichtlineares Regressionsmodell in Form eines Polynommodells getestet, um nichtlineare Muster zwischen dem Merkmal 'Shell weight' und der Zielvariablen 'Rings' zu erfassen. Hierbei wurden Polynomfunktionen unterschiedlicher Grade (2, 3, 4, 5) angewendet, um die beste Passform zu finden.

Die Funktion poly_fit_and_predict bereitet die Daten entsprechend dem Polynomgrad auf, trainiert das lineare Regressionsmodell auf diesen erweiterten Features und berechnet den RMSE, um die Fehlerquote zu bestimmen. Das Modell mit dem niedrigsten RMSE wird als das beste Modell ausgew√§hlt.

Dieser Ansatz hilft, die F√§higkeit des Modells zu verbessern, komplexere Beziehungen zu erkennen, die m√∂glicherweise nicht durch eine einfache lineare Regression erfasst werden. Nach dem Durchlaufen des Prozesses mit den gegebenen Graden, ergab sich der beste Polynomgrad, der auf dem Trainingsset den niedrigsten Fehler aufwies. Dieser Grad repr√§sentiert das Modell, das die beste Balance zwischen Bias und Varianz bietet und ist somit am besten geeignet, die Beziehung zwischen 'Shell weight' und 'Rings' vorherzusagen.

# Multivariate Nonlinear Regression Model

Die Funktion multivariate_poly_fit_and_predict wurde entwickelt, um ein multivariates Polynommodell auf die Trainingsdaten anzuwenden und das Modell zu finden, das den geringsten RMSE auf diesen Daten aufweist. Sie durchl√§uft verschiedene Polynomgrade ‚Äì in diesem Fall Grade 2 bis 5 ‚Äì und berechnet f√ºr jeden Grad ein Polynomfeature, trainiert ein lineares Regressionsmodell und berechnet dessen RMSE. Das Modell mit dem niedrigsten RMSE wird als optimal angesehen.

Nach der Modellwahl wird das beste Modell verwendet, um Vorhersagen auf dem Testdatensatz zu machen. Dabei wird f√ºr das Testset das gleiche Polynomfeature angewendet, das f√ºr das Trainingsset mit dem besten Grad verwendet wurde.

Die Funktion entfernt auch die Zielvariable 'Rings' aus den Trainings- und Testdatens√§tzen, falls diese noch enthalten sein sollte, um eine unvoreingenommene Sch√§tzung zu gew√§hrleisten. Das endg√ºltige Modell, seine Vorhersagen, der beste Grad des Polynoms, der niedrigste Trainingsfehler und die Fehler f√ºr alle getesteten Grade werden dann zur√ºckgegeben.

Dieser Ansatz erm√∂glicht es, die Komplexit√§t des Modells an die Daten anzupassen, indem er eine optimale Balance zwischen Bias und Varianz sucht, um die Genauigkeit der Vorhersagen zu maximieren.







```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error
from math import sqrt
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Angenommen, die Datenframes X_train, X_test und die Zielvariablen y_train, y_test sind bereits vorbereitet.

def poly_regression_analysis(X, y, degrees=[2, 3, 4, 5]):
    error_list = []
    predictions_dict = {}
    for degree in degrees:
        poly_feat_maker = PolynomialFeatures(degree=degree)
        X_poly = poly_feat_maker.fit_transform(X)
        poly_regressor = LinearRegression()
        poly_regressor.fit(X_poly, y)
        predictions = poly_regressor.predict(X_poly)
        model_rmse = sqrt(mean_squared_error(y, predictions))
        error_list.append(model_rmse)
        predictions_dict[degree] = predictions

    return error_list, predictions_dict

# √úberpr√ºfen, ob die Spalte 'Rings' vorhanden ist, und falls ja, entfernen
X_train_adjusted = X_train.drop(columns='Rings', errors='ignore')
X_test_adjusted = X_test.drop(columns='Rings', errors='ignore')

# Durchf√ºhrung der Regression und Vorhersage
train_errors, train_predictions_dict = poly_regression_analysis(X_train_adjusted, y_train, degrees=[2, 3, 4, 5])
test_errors, test_predictions_dict = poly_regression_analysis(X_test_adjusted, y_test, degrees=[2, 3, 4, 5])

# Ausgabe der RMSE-Werte f√ºr das Training
print("Trainingsdaten RMSE:")
for degree, rmse in zip([2, 3, 4, 5], train_errors):
    print(f"Grad {degree}, RMSE: {rmse:.2f}")

# Ausgabe der RMSE-Werte f√ºr den Test
print("\nTestdaten RMSE:")
for degree, rmse in zip([2, 3, 4, 5], test_errors):
    print(f"Grad {degree}, RMSE: {rmse:.2f}")

# Plots hinzuf√ºgen ...

# Bestes Polynommodell auf Trainingsdaten visualisieren
best_degree = np.argmin(train_errors) + 2  # +2, da die Grade bei 2 beginnen
plt.figure(figsize=(10, 6))
plt.scatter(X_train_adjusted['Shell weight'], y_train, color='yellow', alpha=0.7,s=50, label='Tats√§chlich')
plt.scatter(X_train_adjusted['Shell weight'], train_predictions_dict[best_degree], color='purple',alpha=0.7,s=50, label='Vorhergesagt')
plt.title(f'Bester Polynomfit auf Trainingsdaten (Grad {best_degree})')
plt.xlabel('Muschelgewicht')
plt.ylabel('Anzahl der Ringe')
plt.legend()
plt.show()

# Scatterplot von tats√§chlichen vs. vorhergesagten Ringen auf den Testdaten
plt.figure(figsize=(10, 6))
plt.scatter(y_test, test_predictions_dict[best_degree], color='teal', alpha=0.7, s=50, label='Vorhergesagt')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2, label='Ideale Linie')
plt.title(f'Vorhergesagt vs. Tats√§chlich auf Testdaten (Grad {best_degree})', fontsize=14)
plt.xlabel('Tats√§chliche Anzahl der Ringe', fontsize=12)
plt.ylabel('Vorhergesagte Anzahl der Ringe', fontsize=12)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

```
