# Lineare Regression

In der Modellierungsphase unseres Projekts wurde eine lineare Regression implementiert, um den Einfluss des Merkmals 'Shell weight' auf die Zielvariable 'Rings' zu untersuchen. Hierbei wurde der Lineare-Regression-Algorithmus aus sklearn.linear_model verwendet. Dieses Modell ist besonders geeignet für Daten mit linearen Beziehungen zwischen unabhängigen und abhängigen Variablen.

Um die Leistung unseres Modells zu bewerten, wählten wir den Root Mean Squared Error (RMSE) als Metrik. RMSE ist eine gängige und aussagekräftige Metrik für Regressionsmodelle, da sie die durchschnittliche Abweichung der Modellvorhersagen von den tatsächlichen Werten misst und dabei größere Fehler stärker gewichtet. Dies ist besonders relevant bei der Altersvorhersage von Abalonen, wo eine genaue Vorhersage entscheidend ist.

Die Ergebnisse zeigten auf dem Trainingsset einen RMSE von 2.31 und auf dem Testset einen RMSE von 2.43. Diese Werte bedeuten, dass das Modell im Durchschnitt eine Abweichung von ca. 2.31 bzw. 2.43 Ringen von den tatsächlichen Werten aufweist. Die ähnliche Größenordnung des RMSE für beide Datensätze deutet darauf hin, dass unser Modell keine signifikante Über- oder Unteranpassung zeigt und ein gewisses Maß an Generalisierbarkeit auf neue Daten bietet.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from math import sqrt
import matplotlib.pyplot as plt
import numpy as np

# Funktion, um das lineare Modell zu trainieren und Vorhersagen zu machen
def linear_fit(X, y):
    regressor = LinearRegression()
    regressor.fit(X, y)
    y_pred = regressor.predict(X)
    return y_pred, regressor

# Trainieren des Modells und Berechnen der Vorhersagen für das Trainingsset
y_pred_train, model = linear_fit(X_train[['Shell weight']], y_train)


# Berechnen der Vorhersagen für das Testset
y_pred_test, _ = linear_fit(X_test[['Shell weight']], y_test)

# RMSE berechnen
error_train = sqrt(mean_squared_error(y_train, y_pred_train))
error_test = sqrt(mean_squared_error(y_test, y_pred_test))
print("RMSE Error for Linear Regression, Train data is: %.2f" % error_train)
print("RMSE Error for Linear Regression, Test data is: %.2f" % error_test)
```

# Multivariate Lineare Regression

Für eine umfassendere Betrachtung des Zusammenhangs zwischen den Merkmalen und der Zielvariablen wurde eine multivariate lineare Regression durchgeführt. Dieser Ansatz berücksichtigt alle verfügbaren Merkmale gleichzeitig, um die Vorhersagegenauigkeit zu verbessern.

Die Implementierung und das Training des multivariaten Modells erfolgten mithilfe der LinearRegression-Klasse von sklearn. Nach dem Training wurden Vorhersagen für das Trainings- und Testset generiert und der RMSE berechnet, um die Modellleistung zu evaluieren.

Die resultierenden RMSE-Werte von 2.20 für das Trainings- und 2.18 für das Testset zeigen, dass das multivariate Modell eine konsistente Leistung über beide Datensätze hinweg bietet, was auf ein gut generalisierendes Modell hindeutet.
```python
def multiple_fit(X, y):
    regressor = LinearRegression()
    regressor.fit(X, y)
    pred = regressor.predict(X)
    return pred

# Vorhersagen für Trainings- und Testdaten direkt mit X_train, X_test machen
y_pred_train = multiple_fit(X_train, y_train)
y_pred_test = multiple_fit(X_test, y_test)  # Hinweis: Sollte eigentlich nur mit X_train fit gemacht werden

# RMSE für Trainings- und Testdaten berechnen
multiple_error_train = sqrt(mean_squared_error(y_train, y_pred_train))
multiple_error_test = sqrt(mean_squared_error(y_test, y_pred_test))
print("RMSE Error for Multiple Regression, Train data is: %.2f" % multiple_error_train)
print("RMSE Error for Multiple Regression, Test data is: %.2f" % multiple_error_test)
```

# Simple Nonlinear Regression Model

Im nächsten Schritt unseres Projekts wurde ein einfaches nichtlineares Regressionsmodell in Form eines Polynommodells getestet, um nichtlineare Muster zwischen dem Merkmal 'Shell weight' und der Zielvariablen 'Rings' zu erfassen. Hierbei wurden Polynomfunktionen unterschiedlicher Grade (2, 3, 4, 5) angewendet, um die beste Passform zu finden.

Die Funktion poly_fit_and_predict bereitet die Daten entsprechend dem Polynomgrad auf, trainiert das lineare Regressionsmodell auf diesen erweiterten Features und berechnet den RMSE, um die Fehlerquote zu bestimmen. Das Modell mit dem niedrigsten RMSE wird als das beste Modell ausgewählt.

Dieser Ansatz hilft, die Fähigkeit des Modells zu verbessern, komplexere Beziehungen zu erkennen, die möglicherweise nicht durch eine einfache lineare Regression erfasst werden. Nach dem Durchlaufen des Prozesses mit den gegebenen Graden, ergab sich der beste Polynomgrad, der auf dem Trainingsset den niedrigsten Fehler aufwies. Dieser Grad repräsentiert das Modell, das die beste Balance zwischen Bias und Varianz bietet und ist somit am besten geeignet, die Beziehung zwischen 'Shell weight' und 'Rings' vorherzusagen.

```python
def poly_fit_and_predict(X_train, y_train, degrees=[2, 3, 4, 5]):
    # Initialisiere Variablen für das beste Modell und den geringsten Fehler
    best_degree = degrees[0]
    lowest_error = float('inf')
    best_model = None
    best_x_poly_train = None

    # Durchlaufe alle angegebenen Polynomgrade
    for degree in degrees:
        # Transformiere die Eingabedaten in polynomiale Features
        polynomial_features = PolynomialFeatures(degree=degree)
        x_poly_train = polynomial_features.fit_transform(X_train)

        # Trainiere ein lineares Regressionsmodell auf den polynomialen Features
        regressor = LinearRegression()
        regressor.fit(x_poly_train, y_train)
        y_pred_train = regressor.predict(x_poly_train)

        # Berechne den RMSE für das aktuelle Modell
        error = sqrt(mean_squared_error(y_train, y_pred_train))

        # Aktualisiere das beste Modell, wenn der aktuelle Fehler geringer ist
        if error < lowest_error:
            lowest_error = error
            best_degree = degree
            best_model = regressor
            best_x_poly_train = x_poly_train
    
    # Gib das beste Modell und zugehörige Informationen zurück
    return best_model, best_x_poly_train, best_degree, lowest_error

# Trainiere Modelle und wähle das beste Modell basierend auf dem Trainingsset aus
best_model, best_x_poly_train, best_degree, best_train_error = poly_fit_and_predict(
    X_train['Shell weight'].values.reshape(-1, 1), 
    y_train
)

# Vorhersagen für das Testset mit dem besten Modell machen
polynomial_features = PolynomialFeatures(degree=best_degree)
x_poly_test = polynomial_features.fit_transform(X_test['Shell weight'].values.reshape(-1, 1))
y_pred_test = best_model.predict(x_poly_test)
test_error = sqrt(mean_squared_error(y_test, y_pred_test))
```

Das Balkendiagramm stellt den RMSE für verschiedene Polynomgrade dar. Die Stabilität des RMSE über die Grade hinweg lässt darauf schließen, dass komplexere Polynommodelle nicht unbedingt zu besseren Vorhersagen auf den Trainingsdaten führen.



```python
# RMSE für jedes Polynommodell auf dem Trainingsset berechnen und visualisieren
train_errors = []
degrees = [2, 3, 4, 5]
for degree in degrees:
    poly_features = PolynomialFeatures(degree=degree)
    x_poly_train = poly_features.fit_transform(X_train['Shell weight'].values.reshape(-1, 1))
    model = LinearRegression()
    model.fit(x_poly_train, y_train)
    y_poly_pred = model.predict(x_poly_train)
    error = sqrt(mean_squared_error(y_train, y_poly_pred))
    train_errors.append(error)


plt.bar(degrees, train_errors, color='orange', width=0.5)
plt.title("RMSE for Different Polynomial Degrees on Training Data")
plt.xlabel("Degree of polynomial")
plt.ylabel("RMSE")
plt.show()

```

Der RMSE von 2.47 für Trainingsdaten und 2.41 für Testdaten deutet auf eine recht konsistente Vorhersagegenauigkeit des Modells hin, ohne signifikante Über- oder Unteranpassung. Es ist ein Hinweis darauf, dass das Modell die zugrunde liegende Struktur der Daten erfasst, aber bei der Schätzung von Extremwerten nicht so präzise ist, was durch die größere Streuung der Punkte in den Randbereichen des Plots angedeutet wird.

```python
# RMSE für das beste Modell ausgeben
print(f"Best degree: {best_degree} with RMSE: {best_train_error:.2f} on Training Data")
print(f"RMSE: {test_error:.2f} on Test Data with the best polynomial degree of {best_degree}")
```

Der Scatter-Plot illustriert die Ergebnisse eines Polynommodells fünften Grades, angewandt auf den Testdatensatz, um das Alter von Abalonen, repräsentiert durch die Anzahl der Ringe, basierend auf dem Merkmal 'Shell weight' zu schätzen. Die blauen Punkte zeigen die tatsächlichen Alterswerte, während die orangefarbenen Punkte die vom Modell prognostizierten Werte darstellen. Ein idealer Scatter-Plot würde alle Punkte genau auf der diagonalen Linie zeigen, die eine perfekte Übereinstimmung der vorhergesagten mit den tatsächlichen Werten repräsentiert. Die Streuung der Punkte um diese Linie herum deutet darauf hin, dass das Modell gut mit den Daten korreliert, insbesondere in der Mitte des Wertebereichs der 'Rings'.

```python

# Vorhersagen für das Testset mit dem besten Modell machen
polynomial_features = PolynomialFeatures(degree=best_degree)
x_poly_test = polynomial_features.fit_transform(X_test['Shell weight'].values.reshape(-1, 1))
y_pred_test = best_model.predict(x_poly_test)
test_error = sqrt(mean_squared_error(y_test, y_pred_test))

# Visualisiere das beste Modell auf Testdaten
plt.scatter(X_test['Shell weight'], y_test, alpha=0.5, label='Actual')
plt.scatter(X_test['Shell weight'], y_pred_test, alpha=0.5, label='Predicted')
plt.title(f"Polynomial Degree {best_degree} - Test Data")
plt.xlabel("Shell weight")
plt.ylabel("Rings")
plt.legend()
plt.show()
```


Die Säulengrafik veranschaulicht den RMSE für Polynommodelle verschiedener Grade, berechnet auf den Trainingsdaten. # Multivariate Nonlinear Regression Model

Die Funktion multivariate_poly_fit_and_predict wurde entwickelt, um ein multivariates Polynommodell auf die Trainingsdaten anzuwenden und das Modell zu finden, das den geringsten RMSE auf diesen Daten aufweist. Sie durchläuft verschiedene Polynomgrade – in diesem Fall Grade 2 bis 5 – und berechnet für jeden Grad ein Polynomfeature, trainiert ein lineares Regressionsmodell und berechnet dessen RMSE. Das Modell mit dem niedrigsten RMSE wird als optimal angesehen.

Nach der Modellwahl wird das beste Modell verwendet, um Vorhersagen auf dem Testdatensatz zu machen. Dabei wird für das Testset das gleiche Polynomfeature angewendet, das für das Trainingsset mit dem besten Grad verwendet wurde.

Die Funktion entfernt auch die Zielvariable 'Rings' aus den Trainings- und Testdatensätzen, falls diese noch enthalten sein sollte, um eine unvoreingenommene Schätzung zu gewährleisten. Das endgültige Modell, seine Vorhersagen, der beste Grad des Polynoms, der niedrigste Trainingsfehler und die Fehler für alle getesteten Grade werden dann zurückgegeben.

Dieser Ansatz ermöglicht es, die Komplexität des Modells an die Daten anzupassen, indem er eine optimale Balance zwischen Bias und Varianz sucht, um die Genauigkeit der Vorhersagen zu maximieren.







```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error
from math import sqrt
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Angenommen, X_train, X_test, y_train, y_test sind bereits definiert.

def multivariate_poly_fit_and_predict(X_train, y_train, X_test, degrees=[2, 3, 4, 5]):
    best_degree = None
    lowest_error = float('inf')
    best_model = None
    errors = []

    # Trainiere und evaluiere Modelle für jeden Grad.
    for degree in degrees:
        polynomial_features = PolynomialFeatures(degree=degree)
        x_poly_train = polynomial_features.fit_transform(X_train)
        
        regressor = LinearRegression()
        regressor.fit(x_poly_train, y_train)
        y_pred_train = regressor.predict(x_poly_train)
        
        error = sqrt(mean_squared_error(y_train, y_pred_train))
        errors.append(error)
        if error < lowest_error:
            lowest_error = error
            best_degree = degree
            best_model = regressor

    # Verwende das beste Modell, um das Testset vorherzusagen.
    x_poly_test = polynomial_features.fit_transform(X_test)
    y_pred_test = best_model.predict(x_poly_test)
    
    return best_model, y_pred_test, best_degree, lowest_error, errors

# Die Funktion sollte nach ihrer Definition aufgerufen werden.
# Überprüfen, ob die 'Rings'-Spalte vorhanden ist, und wenn ja, entferne sie.
X_train_prepared = X_train.drop('Rings', axis=1) if 'Rings' in X_train.columns else X_train
X_test_prepared = X_test.drop('Rings', axis=1) if 'Rings' in X_test.columns else X_test

# Aufruf der Funktion mit den vorbereiteten DataFrames.
best_model, y_pred_test, best_degree, best_train_error, train_errors = multivariate_poly_fit_and_predict(
    X_train_prepared, 
    y_train,
    X_test_prepared,
    degrees=[2, 3, 4, 5]
)
```
